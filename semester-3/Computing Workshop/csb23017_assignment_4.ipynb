```python
#1

def add_new(dictionary, key, value):
    dictionary[key] = value;
    print("The updated dictionary : ",dictionary);

my_dict = {
    'name' : 'Raktim Ranjan Handique',
    'dept' : 'Computer Science and Engineering',
}
add_new(my_dict,'Roll no', 'csb23017')
    
```

```python
#2

def filter_dict(dictionary, threshold):
    filtered_dict = {}
    for key, value in dictionary.items():
        if value>threshold:
            filtered_dict[key] = value;
    return filtered_dict;

my_dict = {
    'x' : 100,
    'y' : 200,
    'z' : 300,
}
result = filter_dict(my_dict, 150)
print("The filtered dictionary is :",result) 
    
```

```python
#3

def remove_dict(dictionary):
    if dictionary:
        dictionary.popitem()
        print("The updated dictionary is :", dictionary)
    else:
        print("The dictionary is empty.");

my_dict = {
    'name' : 'Raktim Ranjan Handique',
    'dept' : 'Computer Science and Engineering',
    'roll no' : 'csb23017'
}
remove_dict(my_dict)
```

```python
#4

def find_palindromes():
    
    strings = input("Enter a list of strings separated by spaces: ").split()
    palindromes = []
    
    for s in strings:
        if s == s[::-1]:  
            palindromes.append(s)
    
    if palindromes:
        return palindromes
    else:
        return "No palindrome"

result = find_palindromes()
print(result)
```

```python
#4

def char_frequency():
    string = input("Enter a string: ")
    
    frequency_dict = {}  
    for char in string:
        frequency_dict[char] = frequency_dict.get(char, 0) + 1  
    
    return frequency_dict

result = char_frequency()
print(result)
```
