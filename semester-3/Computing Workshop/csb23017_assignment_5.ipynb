```python
#Q1

def find_unique_pair(arr, target_sum):
    pairs = set();
    appeared_numbers = set();

    for num in arr:
        complement = target_sum - num
        if complement in appeared_numbers:
            pairs.add(tuple(sorted((num, complement))))
        appeared_numbers.add(num)
    return list(pairs)

arr = [1, 8, 2, 7, 9, 0, 9, 6]
target_sum = 8
print(find_unique_pair(arr, target_sum))
```

```python
#2

def zero_count(arr):
    count = 0
    result = []

    for num in arr:
        if num == 0:
            count = count+1
        else :
            result.append(num)

    result.extend([0]*count)
    return result, count

arr = [1,4,0,3,0,5,0,55]
result, count = zero_count(arr)
print("Array after moving zeros to the right :",result)
print("Number of zeros :",count)
```

```python
#3 

def intersection_arr(arr1, arr2):
    set1 = set(arr1)
    set2 = set(arr2)

    intersection = set1.intersection(set2)

    return list(intersection)

arr1 = [1,3,5,7,9,10]
arr2 = [2,4,6,8,10]
print(intersection_arr(arr1,arr2))
    
```
