```python
#2


n = int(input("Enter the number of elements in tuple: "))
tup = tuple(int(input(f"Enter element {i+1}: ")) for i in range(n))

min_val = tup[0]
max_val = tup[0]

for i in tup:
    if i < min_val:
        min_val = i
    if i > max_val:
        max_val = i

print(f"Minimum value in the tuple: {min_val}")
print(f"Maximum value in the tuple: {max_val}")
```

```python
#3.a 


m = int(input("Enter the number of elements in T1: "))
T1 = tuple(int(input(f"Enter element {i+1} for T1: ")) for i in range(m))

n = int(input("Enter the number of elements in T2: "))
T2 = tuple(int(input(f"Enter element {i+1} for T2: ")) for i in range(n))

T3 = T1 + T2

list_count = []

for element in T3:
    count = 0
    for x in T3:
        if element == x:
            count += 1
    counts.append(list_count)

print(f"Merged tuple T3: {T3}")
print(f"Occurrences of each element in T3: {counts}")
```

```python
#3.b

m = int(input("Enter the number of elements in T1: "))
T1 = tuple(int(input(f"Enter element {i+1} for T1: ")) for i in range(m))

n = int(input("Enter the number of elements in T2: "))
T2 = tuple(int(input(f"Enter element {i+1} for T2: ")) for i in range(n))

T3 = T1 + T2

T3_unique_sorted = sorted(set(T3))

print(f"Sorted tuple with duplicates removed: {T3_unique_sorted}")
```
