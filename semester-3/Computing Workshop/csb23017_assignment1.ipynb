```python
# question 1

price = 1000;
x = 1000*(20/100);
y = price - x;
price1 = y;
print("The price after 1st discount is ",price1);

z = (10/100)*price1;
price2 = price1 - z;
print("The price after 2nd discount is ",price2);

final_price = price2;
final_discount = ((z+x)/price)*100;
print("The overall discount of the shirt is ",final_discount,"%");
```

```python
# question 2b

m = int(input("Enter how much money Mr. Jones has :"));

item1 = 50.78 ;
item2 = 100.89 ;
x = int(input("Enter the number of item 1 : \n"));
print(x);
y = int(input("Enter the number of item 2 : \n"));
print(y);

price = x*item1 + y*item2;
print("The price before tax is :",price);

z = price*(18/100);
final_price = price + z;
print("The final amount should be paid is :",final_price);

left = m - final_price;
print("money left with Mr. John is :",left);
```
