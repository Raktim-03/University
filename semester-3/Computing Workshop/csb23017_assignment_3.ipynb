```python
#1
def add_new():
    dictionary = eval(input("Enter the dictionary : "))                   #like {'name' : 'kishan', 'designation' : 'topper'}
    
    key = input("Enter the key to add: ")
    value = input("Enter the value for the key: ")
    
    dictionary[key] = value
    print("Updated dictionary:", dictionary)

add_new()
```

```python
#2

def filter_dict(dictionary, threshold):
    filtered_dict = {}
    for key, value in dictionary.items():
        if value>threshold:
            filtered_dict[key] = value;
    return filtered_dict;

my_dict = {
    'x' : 100,
    'y' : 200,
    'z' : 300,
}
result = filter_dict(my_dict, 150)
print("The filtered dictionary is :",result) 
```

```python
#3

def remove_dict(dictionary):
    if dictionary:
        dictionary.popitem()
        print("The updated dictionary is :", dictionary)
    else:
        print("The dictionary is empty.");

my_dict = {
    'name' : 'Raktim Ranjan Handique',
    'dept' : 'Computer Science and Engineering',
    'roll no' : 'csb23017'
}
remove_dict(my_dict)
```
