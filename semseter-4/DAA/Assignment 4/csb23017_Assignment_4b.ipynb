```python
def read_students_from_file(filename):
    with open(filename, 'r') as file:
        return sorted([(line.split()[0], int(line.split()[1]), int(line.split()[2])) for line in file])

def Sort(students, sub):
    if sub not in ["biology", "maths"]:
        raise ValueError("Invalid subject. Choose 'biology' or 'maths'.")
    index = 1 if sub == "biology" else 2
    for s in sorted(students, key=lambda x: x[index], reverse=True):
        print(s[0], s[index])

def printList(students, sub, min_marks, max_marks):
    if sub not in ["biology", "maths"]:
        raise ValueError("Invalid subject. Choose 'biology' or 'maths'.")
    index = 1 if sub == "biology" else 2
    sorted_students = sorted(students, key=lambda x: x[index], reverse=True)
    marks = [s[index] for s in sorted_students]
    left, right = 0, len(marks) - 1
    start, end = -1, -1
    while left <= right:
        mid = (left + right) // 2
        if marks[mid] >= min_marks:
            start = mid
            right = mid - 1
        else:
            left = mid + 1
    left, right = 0, len(marks) - 1
    while left <= right:
        mid = (left + right) // 2
        if marks[mid] <= max_marks:
            end = mid
            left = mid + 1
        else:
            right = mid - 1
    if start != -1 and end != -1:
        for i in range(start, end + 1):
            print(sorted_students[i][0], sorted_students[i][index])

def categorize_students(students):
    sorted_maths = sorted(students, key=lambda x: x[2], reverse=True)
    sorted_bio = sorted(students, key=lambda x: x[1], reverse=True)
    rank_math = {s[0]: i + 1 for i, s in enumerate(sorted_maths)}
    rank_bio = {s[0]: i + 1 for i, s in enumerate(sorted_bio)}
    for s in sorted(students):
        decision = "ENGG" if rank_math[s[0]] < rank_bio[s[0]] else "MED" if rank_math[s[0]] > rank_bio[s[0]] else "ANY"
        print(s[0], decision)

if __name__ == "__main__":
    students = read_students_from_file("marks.txt")
    print("Sorted by Biology:")
    Sort(students, "biology")
    print("\nStudents with Biology marks between 70 and 90:")
    printList(students, "biology", 70, 90)
    print("\nPrincipal's Decision:")
    categorize_students(students)
```
