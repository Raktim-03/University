```python
import random

def generate_players(n):
    players = random.sample(range(1, 101), n)
    print("Player Strengths:", players)
    return players

def tournament(players):
    rounds = [players[:]]
    matchups = []
    while len(players) > 1:
        next_round = []
        current_matchups = []
        for i in range(0, len(players), 2):
            if i + 1 < len(players):
                winner = max(players[i], players[i + 1])
                current_matchups.append((players[i], players[i + 1], winner))
            else:
                winner = players[i]
                current_matchups.append((players[i], "BYE", winner))
            next_round.append(winner)
        matchups.append(current_matchups)
        players = next_round
        rounds.append(players[:])
    print_matches(matchups)
    return rounds

def print_matches(matchups):
    for round_num, matches in enumerate(matchups, start=1):
        print(f"Round {round_num}:")
        for match in matches:
            if match[1] == "BYE":
                print(f"{match[0]} automatically advances")
            else:
                print(f"{match[0]} vs {match[1]} -> Winner: {match[2]}")
        print()

if __name__ == "__main__":
    num_players = 8
    players = generate_players(num_players)
    tournament_rounds = tournament(players)
```
