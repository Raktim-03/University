```python
def construct_enclosure(points):
    if not points:
        return [(0, 0)]
    
    points = sorted(points, key=lambda p: (p[0], p[1]))
    enclosure = [(0, 0), (points[0][0], 0)]
    
    prev_x, prev_y = points[0][0], 0
    for x, y in points:
        if y != prev_y:
            enclosure.append((prev_x, y))
            prev_y = y
        if x != prev_x:
            enclosure.append((x, prev_y))
            prev_x = x
    
    enclosure.append((points[-1][0], points[-1][1]))
    top_y = max(y for x, y in points)
    enclosure.append((points[-1][0], top_y))
    
    points = sorted(points, key=lambda p: (-p[0], -p[1]))
    prev_x, prev_y = points[0][0], top_y
    for x, y in points:
        if y != prev_y:
            enclosure.append((prev_x, y))
            prev_y = y
        if x != prev_x:
            enclosure.append((x, prev_y))
            prev_x = x
    
    enclosure.append((0, enclosure[-1][1]))
    enclosure.append((0, 0))
    
    return enclosure

def main():
    points = [(1,1), (2,3), (3,2), (4,4), (2,1)]
    for x, y in construct_enclosure(points):
        print(f"({x},{y})")

if __name__ == "__main__":
    main()
```
