```python
# question a.

def IsSummable(Z, A, index=0):
    if index == len(A):
        return Z == 0
    return (
        IsSummable(Z - A[index], A, index + 1) or
        IsSummable(Z + A[index], A, index + 1) or
        IsSummable(Z, A, index + 1)
    )

A = list(map(int, input("Elements you want to add in the array: ").split()))
Z = int(input("Target value: "))

if IsSummable(Z, A):
    print("Yes, Summabale")
else:
    print("No, Not Summable")
```

```python
#question b

def CountBall(h):
    if h == 1:
        return 1
    return h * h + CountBall(h - 1)

h = int(input("Enter the height of the pyramid: "))
print("Total number of balls:", CountBall(h))
```

```python
import random

def ask(a, b):
    if people[a] == 1:
        return people[b]
    else:
        return random.choice([0, 1])

def find_one_honest():
    for i in range(len(people)):
        count = 0
        for j in range(len(people)):
            if i != j and ask(i, j) == 1:
                count += 1
        if count > len(people) // 2:
            return i
    return -1

def find_all_honest(honest_person):
    honest_list = []
    for i in range(len(people)):
        if i == honest_person or ask(honest_person, i) == 1:
            honest_list.append(i)
    return honest_list

n = int(input("Enter number of people in the club: "))
people = [1]*((n//2)+1) + [0]*(n - ((n//2)+1))
random.shuffle(people)

honest = find_one_honest()
if honest != -1:
    all_honest = find_all_honest(honest)
    print("One honest person:", honest)
    print("All honest people:", all_honest)
else:
    print("Could not determine an honest person.")
```
