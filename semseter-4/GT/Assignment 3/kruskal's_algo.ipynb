```python
edges = [
    (14, 13, 70), (14, 8, 230), (14, 7, 210), (14, 6, 300), (14, 4, 400),
    (14, 15, 80), (14, 19, 350), (4, 5, 80), (4, 6, 180), (4, 3, 350),
    (4, 0, 550), (4, 10, 350), (0, 2, 300), (2, 1, 100), (2, 3, 230),
    (6, 7, 110), (7, 8, 210), (10, 12, 150), (12, 15, 230), (0, 9, 220),
    (9, 11, 150), (11, 16, 400), (16, 17, 110), (16, 18, 400), (18, 19, 500),
    (17, 19, 400), (19, 20, 130), (20, 21, 210), (21, 22, 300)
]

edges.sort(key=lambda x: x[2])
parent = list(range(23))

def find(u):
    if parent[u] != u:
        parent[u] = find(parent[u])
    return parent[u]

mst = []
for u, v, w in edges:
    if find(u) != find(v):
        parent[find(v)] = find(u)
        mst.append((u, v, w))
        if len(mst) == 22:
            break

for edge in mst:
    print(f"{edge[0]} - {edge[1]}: {edge[2]}m")
print("Total:", sum(w for _, _, w in mst), "m")
```
