```python
import heapq

edges = [
    (14, 13, 70), (14, 8, 230), (14, 7, 210), (14, 6, 300), (14, 4, 400),
    (14, 15, 80), (14, 19, 350), (4, 5, 80), (4, 6, 180), (4, 3, 350),
    (4, 0, 550), (4, 10, 350), (0, 2, 300), (2, 1, 100), (2, 3, 230),
    (6, 7, 110), (7, 8, 210), (10, 12, 150), (12, 15, 230), (0, 9, 220),
    (9, 11, 150), (11, 16, 400), (16, 17, 110), (16, 18, 400), (18, 19, 500),
    (17, 19, 400), (19, 20, 130), (20, 21, 210), (21, 22, 300)
]

graph = [[] for _ in range(23)]
for u, v, w in edges:
    graph[u].append((v, w))
    graph[v].append((u, w))

visited = [False] * 23
heap = [(0, 0, -1)]
mst = []

while heap:
    w, u, parent = heapq.heappop(heap)
    if not visited[u]:
        visited[u] = True
        if parent != -1:
            mst.append((parent, u, w))
        for v, edge_w in graph[u]:
            if not visited[v]:
                heapq.heappush(heap, (edge_w, v, u))

for edge in mst:
    print(f"{edge[0]} - {edge[1]}: {edge[2]}m")
print("Total:", sum(e[2] for e in mst), "m")
```
